### 04) WAP to find the longest word in a file named abc.txt

```python
max=" "
text_file=open('abc.txt')
lines=text_file.readlines();
for line in lines:
    words=line.split(" ")
    for word in words:
        if len(word)>len(max):
            max=word
print(f"longest word:~ {max}")
text_file.close() 
```

### 05) WAP to find the size of the file named abc.txt

```python
a=open("abc.txt",'r')
count=0
k=a.readlines()
for i in k:
    words=i.split()
    for w in words:
        for j in w:
            count+=1
print(count)
a.close()
```

### 06) WAP to implement search function to search specific occurance of word in a given text file.

```python
def search(file_path,word):
    text_file=open(file_path,'r')
    content=text_file.read()
    if word in content:
        print('STRING EXIST IN A FILE :',word)
    else:
        print('STRING DOES NOT EXIST IN A FILE')
        
search('abc.txt','Imagination')
search('abc.txt','Darshan')
```

### 01) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
text_file=open('primenumbers.txt','w')
for m in range(2,100):
    if all(m%i!=0 for i in range(2,m)):
        text_file.write(str(m))
        text_file.write('\n')
text_file.close()
```

### 03) WAP to encrypt a text file.

```python
f=open('ref1.txt','w')
a=input('ENTER A STRING IN ALPHABETS:')
key=int(input('ENTER A KEY:'))
for i in range(len(a)):
    f.write(chr(ord(a[i])+key))
f.close()
```

### 04) WAP to decrypt a previously encrypted file.

```python
f=open('ref1.txt','w')
a=input('ENTER A STRING IN ALPHABETS:')
key=int(input('ENTER A KEY:'))
for i in range(len(a)):
    f.write(chr(ord(a[i])-key))
f.close()
```
