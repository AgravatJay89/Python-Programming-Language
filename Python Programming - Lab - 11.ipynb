```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]


# write a code to generate below graph
plt.plot(x,cxMarks,color='r',ls='--',marker='o')
plt.plot(x,cyMarks,color='b',ls=':',marker='>')
plt.xlabel('roll no')
plt.ylabel('cpi')
plt.annotate(xy=[8,1],text='Lowest Spi')
plt.legend(['cx','cy'],loc=3)
# plt.show()
plt.savefig('MyChart.jpg')
```

### 01) WAP to demonstrate the use of Pie chart.

```python
x = 10,20,30,40
y = 'food','travel','accomodation','shopping'
c = ['r','g','y','m']
ex=[0.5,0.0,0.0,0.0]
plt.pie(x,labels=y,colors=c,shadow=True,radius=1.3,autopct="%0.1f%%",labeldistance=1.2,textprops={"fontsize":15,"color":'r',},explode=ex)
plt.title("Expenses")
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x = 'python','c','c++','java'
y = 85,70,60,80
c = ['r','g','y','m']

plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.5,)
plt.title('Popularity of language')
plt.xlabel('Language')
plt.ylabel('Popularity')
plt.show()
```

```python
x = 'python','c','c++','java'
y = 85,70,60,80
c = ['r','g','y','m']

plt.barh(x,y,color=c,align='edge',edgecolor='b',alpha=0.5,)
plt.title('Popularity of language')
plt.xlabel('Popularity')
plt.ylabel('Language')
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = 'python','c','c++','java'
y = 85,70,60,80
c = ['r','g','y','m']

plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.5,)
plt.title('Popularity of language')
plt.xlabel('Language')
plt.ylabel('Popularity')

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])

addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = 1,2,3,4,5
y = 2,4,6,8,10
c = ['r','g','y','m','b']
s = [100,200,300,400,500]
plt.scatter(x,y,color=c,marker='*',s=s,edgecolors='b')
plt.title('Scatter Plot')
plt.xlabel('X -axis')
plt.ylabel('Y - axis')
plt.show()
```
